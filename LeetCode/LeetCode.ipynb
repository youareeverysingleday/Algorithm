## <a id='toc1_3_'></a>[1. 两数之和](#toc0_)

给定一个整数数组 nums 和一个整数目标值 target，请你在该数组中找出 和为目标值 target  的那 两个 整数，并返回它们的数组下标。

你可以假设每种输入只会对应一个答案，并且你不能使用两次相同的元素。

你可以按任意顺序返回答案。

- 示例 1：
    ``` log
    输入：nums = [2,7,11,15], target = 9
    输出：[0,1]
    解释：因为 nums[0] + nums[1] == 9 ，返回 [0, 1] 。
    ```
- 示例 2：

    ```log
    输入：nums = [3,2,4], target = 6
    输出：[1,2]
    ```
- 示例 3：
    ```log
    输入：nums = [3,3], target = 6
    输出：[0,1]
    ```
- 提示：
  - $2 <= nums.length <= 10^4$
  - $-10^9 <= nums[i] <= 10^9$
  - $-10^9 <= target <= 10^9$
  - 只会存在一个有效答案

进阶：你可以想出一个时间复杂度小于 $O(n^2)$ 的算法吗？

### <a id='toc1_3_1_'></a>[解答](#toc0_)

1. 暴力查找法

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        listlen = len(nums)
        i = 0
        j = 0
        flag = 0
        for i in range(listlen):
            n0 = nums[i]
            for j in range(i+1, listlen, 1):
                n1 = nums[j]
                if n0 + n1 == target:
                    flag = 1
                    break
            if flag == 1:
                break

        return [i, j]
```

### <a id='toc1_5_1_'></a>[常规算法](#toc0_)

中位数的定义
- 若合并后的长度为奇数：中位数是中间的那个元素。
- 若合并后的长度为偶数：中位数是中间两个元素的平均值。

常规思路（合并+排序）
- 合并两个数组 → $O(m+n)$。Python的list合并操作是新创建一个List，然后将需要合并的两个list中的元素组个合并到新创建的list中。
- 排序 → $O((m+n)log_2(m+n))$
- 取中位数 → $O(1)$

算法复杂度 $O(m+n) + O((m+n)log_2 (m+n)) + O(1)$

````python
from typing import List

class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        result = nums1 + nums2
        result.sort()
        length = len(result)
        if length % 2 == 0:
            return (result[length//2 - 1] + result[length//2]) / 2
        else:
            return result[length//2]    
````

### <a id='toc1_5_2_'></a>[二分法](#toc0_)

优化思路：二分查找
- 可以用二分法在两个有序数组中查找第 k 小的数。
- 要求中位数，其实就是查找合并后第 (m + n) // 2 + 1 小的数。

思路：
- 将问题转换为求第 k 小的数。
- 在两个数组中用二分法，每次丢弃 k/2 个元素（在较小的那一段中）。
- 通过缩小搜索范围，最终定位到第 k 小的元素。

````python
def findMedianSortedArrays(nums1, nums2):
    def get_kth_element(k):
        index1, index2 = 0, 0
        
        while True:
            # 如果一个数组已经遍历完，直接返回另一个数组的第 k 小元素
            if index1 == len(nums1):
                return nums2[index2 + k - 1]
            if index2 == len(nums2):
                return nums1[index1 + k - 1]
            # 如果 k = 1，返回两个数组当前指针中较小的元素
            if k == 1:
                return min(nums1[index1], nums2[index2])
            
            # 正常情况下，每次比较第 k//2 个元素
            new_index1 = min(index1 + k // 2 - 1, len(nums1) - 1)
            new_index2 = min(index2 + k // 2 - 1, len(nums2) - 1)
            pivot1, pivot2 = nums1[new_index1], nums2[new_index2]
            
            if pivot1 <= pivot2:
                k -= (new_index1 - index1 + 1)
                index1 = new_index1 + 1
            else:
                k -= (new_index2 - index2 + 1)
                index2 = new_index2 + 1
    
    n = len(nums1) + len(nums2)
    
    if n % 2 == 1:
        # 奇数长度，返回第 n//2 + 1 个元素
        return get_kth_element(n // 2 + 1)
    else:
        # 偶数长度，返回中间两个元素的平均值
        return (get_kth_element(n // 2) + get_kth_element(n // 2 + 1)) / 2.0

# 测试
nums1 = [1, 3]
nums2 = [2]
print(findMedianSortedArrays(nums1, nums2))  # 输出: 2.0

nums1 = [1, 2]
nums2 = [3, 4]
print(findMedianSortedArrays(nums1, nums2))  # 输出: 2.5
````

给你两个单链表的头节点 headA 和 headB ，请你找出并返回两个单链表相交的起始节点。如果两个链表不存在相交节点，返回 null 。


题目数据 保证 整个链式结构中不存在环。

注意，函数返回结果后，链表必须 保持其原始结构 。

自定义评测：

评测系统 的输入如下（你设计的程序 不适用 此输入）：
- intersectVal - 相交的起始节点的值。如果不存在相交节点，这一值为 0
- listA - 第一个链表
- listB - 第二个链表
- skipA - 在 listA 中（从头节点开始）跳到交叉节点的节点数
- skipB - 在 listB 中（从头节点开始）跳到交叉节点的节点数
评测系统将根据这些输入创建链式数据结构，并将两个头节点 headA 和 headB 传递给你的程序。如果程序能够正确返回相交节点，那么你的解决方案将被 视作正确答案 。

````python

# Definition for singly-linked list.
from typing import Optional

class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        if not headA or not headB:
            return None
        
        for nodeA in headA:
            for nodeB in headB:
                if nodeA == nodeB:
                    return nodeA

        
        return 0
````

## <a id='toc1_10_'></a>[226. 翻转二叉树](#toc0_)

````python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        

class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        if not root:
            return None
        root.left, root.right = self.invertTree(root.right),  self.invertTree(root.left)  # 交换左右子树
        return root
````

- 如果孩子找到答案是 node，那当前节点的答案还是它；
- 如果没找到，但 p 和 q 分别位于左右孩子，那当前节点就是答案；
- 如果当前节点是其中一个 p 或 q，那只能返回当前这个，表示最近的是本身；
- 如果 p 和 q 位于同一侧孩子处，本质上就是找到答案。

````python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

root = TreeNode(3)
root.left = TreeNode(5)
root.right = TreeNode(1)
root.left.left = TreeNode(6)
root.left.right = TreeNode(2)
root.right.left = TreeNode(0)
root.right.right = TreeNode(8)
root.left.right.left = TreeNode(7)
root.left.right.right = TreeNode(4)

class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        # 如果 root 为空，或者 root 就是 p 或 q，则 root 就是最近公共祖先
        if not root or root == p or root == q:
            return root

        # 在左子树中查找 LCA（最近公共祖先）。
        left = self.lowestCommonAncestor(root.left, p, q)
        # 在右子树中查找 LCA
        right = self.lowestCommonAncestor(root.right, p, q)

        # 如果左右子树分别找到了 p 和 q，说明当前 root 就是最近公共祖先
        if left and right:
            return root

        # 否则，返回找到的非空子树
        return left if left else right
    
p = root.left  # 节点 5
q = root.right # 节点 1

sl = Solution()

lca = sl.lowestCommonAncestor(root, p, q)
print(lca.val)  # 
````

### <a id='toc1_13_1_'></a>[遍历加栈（栈用于模拟回溯）解法](#toc0_)

````python
from typing import List

class Solution:
    def solveNQueens(self, n: int) -> List[List[str]]:
        result = []
        stack = []  # 使用栈模拟递归
        row, col = 0, 0  # 从第一行、第一列开始

        while True:
            while col < n:
                # 检查 (row, col) 是否安全
                if all(c != col and abs(r - row) != abs(c - col) for r, c in stack):
                    stack.append((row, col))  # 放置皇后
                    row += 1  # 进入下一行
                    col = 0  # 从新的一行开始尝试
                    break
                col += 1  # 尝试下一列

            if len(stack) == n:  # 找到一个解
                result.append(["." * c + "Q" + "." * (n - c - 1) for _, c in stack])
            
            if col == n or len(stack) == n:  # 需要回溯
                if not stack:  # 栈空了，搜索完毕
                    break
                row, col = stack.pop()  # 取出上一行皇后的位置
                col += 1  # 尝试下一列

        return result

# 测试
solution = Solution()
solutions = solution.solveNQueens(5)
print(len(solutions))

````

### <a id='toc1_13_3_'></a>[回溯加递归解法（最简洁解）](#toc0_)

下面有2个答案，一个是Leetcode中填写的，代码不完整。另一个是可以在独立环境中运行的，代码完整。

- board 是一个整数列表，board[i] 存储第 i 行皇后所在的列索引。
- 例如，如果 board = [1, 3, 0, 2]，表示：
  - 第 0 行的皇后放在 1 号列 (Q 在 (0,1))
  - 第 1 行的皇后放在 3 号列 (Q 在 (1,3))
  - 第 2 行的皇后放在 0 号列 (Q 在 (2,0))
  - 第 3 行的皇后放在 2 号列 (Q 在 (3,2))

重点在于建立起来的board，它非常的精妙，它包含了3层含义：
1. board中元素的位置表示了真实棋盘的行号。
2. board中元素的值表示了真实棋盘的列号。
3. **board是一个一维链表，每行只能放一个Q！**所以在is_valid()函数中不用判断同一行是否还有其他的Q。
4. 减少的遍历的次数。每次只有需要查询已经放置了Q的位置，后面的不用管。也就是这句的含义：for i in range(row):

**回溯实际上是利用了代码的一个特点，执行完当前函数之后范围执行函数时的位置。**这样就可以不用记录查找的位置而遍历所有的情况。

````python
class Solution:
    def solveNQueens(self, n: int) -> List[List[str]]:
        # 判断当前位置是否能放置皇后。
        def is_valid(board, row, col):
            # 只遍历已经放置了皇后的行，后面的行不用管。
            for i in range(row):
                # 因为board建立的时候就只允许每行放置一个Q，所以不用判断同行的情况。从而减少循环次数。
                # 一定要注意这里的判断条件是col - row，而不是row - col。这会直接导致结果错误。
                # 因为board[i]对应的是col的值。
                if board[i] == col or \
                board[i] - i == col - row or \
                board[i] + i == col + row:
                    return False
            return True

        # 回溯的意思就是从最后结束的地方在返回到一个循环结束的地方。
        # 使用回溯的优势：1. 会回到之前结束的地方，不用标记这个地方。
        # 2. 只要按照判断条件就可以记录完整的正确答案。
        def backtrack(row):
            if row == n:
                # 这里是在保存回溯完整个行之后找寻到的一个结果。
                result.append(["." * c + "Q" + "." * (n - c - 1) for c in board])
                return
            
            for col in range(n):
                if is_valid(board, row, col):
                    # 把列的位置赋值给board对应行的值。
                    board[row] = col
                    # 这里是在递归调用。
                    backtrack(row + 1)
                    # 在之前回溯完成之后，需要恢复board中已经赋值的位置。方便找寻之前条件确定，而之后变化的情况。
                    board[row] = -1
        
        # 保存所有棋盘的结果。
        result = []
        # 初始化棋盘。
        board = [-1] * n
        # 开始回溯。从0行开始。
        backtrack(0)
        # 返回所有可能的棋盘结果。
        return result
        
````

````python
from typing import List

def solveNQueens(n: int) -> List[List[str]]:
    def is_valid(board, row, col):
        for i in range(row):
            if board[i] == col or \
               board[i] - i == col - row or \
               board[i] + i == col + row:
                return False
        return True

    def backtrack(row):
        if row == n:
            result.append(["." * c + "Q" + "." * (n - c - 1) for c in board])
            return
        
        for col in range(n):
            if is_valid(board, row, col):
                board[row] = col
                backtrack(row + 1)
                board[row] = -1
    
    result = []
    board = [-1] * n
    backtrack(0)
    return result

# 测试
n = 4
solutions = solveNQueens(n)
for sol in solutions:
    for row in sol:
        print(row)
    print('\n')
print(len(solutions))
````

## <a id='toc1_14_'></a>[95. 不同的二叉搜索树](#toc0_)

给你一个整数 n ，求恰由 n 个节点组成且节点值从 1 到 n 互不相同的 二叉搜索树 有多少种？返回满足题意的二叉搜索树的种数。

### <a id='toc1_14_1_'></a>[解题思路](#toc0_)
1. BST 性质：
   1. 左子树的所有节点值 小于 根节点。
   2. 右子树的所有节点值 大于 根节点。
2. 递归构建 BST：
   1. 选取 i（从 1 到 n）作为根节点。
   2. 递归生成 1 到 i-1 作为左子树的所有可能情况。
   3. 递归生成 i+1 到 n 作为右子树的所有可能情况。
   4. 组合左右子树形成完整的 BST。
3. 边界情况：
   1. 若 start > end，返回 [None]（空树）。
   2. 若 start == end，返回 [TreeNode(start)]（单个节点）。


### <a id='toc1_14_2_'></a>[代码解析](#toc0_)
1. build_trees(start, end)：
   1. 递归构建从 start 到 end 的所有可能的 BST。
   2. 终止条件：start > end 时，返回 [None]（空树）。
   3. 遍历 i 作为根节点，递归构造左子树和右子树。
   4. 左右子树的所有组合 都会成为不同的 BST。

### <a id='toc1_14_3_'></a>[时间复杂度](#toc0_)
1. 生成不同 BST 需要递归遍历所有可能性，时间复杂度接近 $O(4^n / sqrt(n))$（卡特兰数）。
2. 适用于 n ≤ 8-10，较大的 n 计算量会指数增长。

### <a id='toc1_14_4_'></a>[总结](#toc0_)
1. 递归 + 回溯 生成所有 BST。
2. 左子树、右子树分别递归生成，然后组合。
3. 复杂度高，适用于小规模 n（n ≤ 10）。

如果 n 比较大，可以用 动态规划 方法优化，只计算 G(n)（不同 BST 数量）。

````python
from typing import List, Optional

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def generateTrees(self, n: int) -> List[Optional[TreeNode]]:
        if n == 0:
            return []
        
        def build_trees(start, end):
            if start > end:
                return [None]
            
            all_trees = []
            for i in range(start, end + 1):  # 选取 i 作为根节点
                left_trees = build_trees(start, i - 1)  # 递归生成左子树
                right_trees = build_trees(i + 1, end)  # 递归生成右子树
                
                # 组合左右子树
                for left in left_trees:
                    for right in right_trees:
                        root = TreeNode(i)
                        root.left = left
                        root.right = right
                        all_trees.append(root)
            
            return all_trees
        
        return build_trees(1, n)


sol = Solution()
trees = sol.generateTrees(3)

# 打印 BST 结构
def preorder(root):
    return [root.val] + preorder(root.left) + preorder(root.right) if root else []

for tree in trees:
    print(preorder(tree))
````

## <a id='toc1_15_'></a>[96. 不同的二叉搜索树的数量](#toc0_)

给你一个整数 n ，求恰由 n 个节点组成且节点值从 1 到 n 互不相同的 二叉搜索树 有多少种？返回满足题意的二叉搜索树的种数。

**解题思路**
这个问题是典型的 **卡特兰数（Catalan Number）** 问题，可以使用动态规划高效求解。

**状态转移方程**
设 `dp[i]` 表示 **由 `i` 个节点可以组成的不同二叉搜索树（BST）的数量**，状态转移方程如下：
$$
dp[i] = \sum_{j=1}^{i} dp[j-1] \times dp[i-j]
$$
其中：
- `j` 作为根节点时，左子树有 `j-1` 个节点，右子树有 `i-j` 个节点。
- 左右子树的构造是独立的，因此总方案数是它们的乘积。

**初始条件**
- `dp[0] = 1`（空树）
- `dp[1] = 1`（只有一个节点）


````python
class Solution:
    def numTrees(self, n: int) -> int:
        dp = [0] * (n + 1)
        dp[0], dp[1] = 1, 1

        for i in range(2, n + 1):
            for j in range(1, i + 1):
                dp[i] += dp[j - 1] * dp[i - j]

        return dp[n]
````
