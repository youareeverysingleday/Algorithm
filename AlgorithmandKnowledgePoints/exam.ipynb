```python
import re
def extract_number(s):
    if not isinstance(s, str):
        return np.nan
    match = re.match(r'^(-?\d+)(?=\[)', s)  # 匹配 [ 前的负数/正数
    if match:
        return int(match.group(1))
    match = re.match(r'^-?\d+$', s)  # 整个字符串是负数/正数
    if match:
        return int(match.group(0))
    return np.nan
```

```python
import pandas as pd
import numpy as np

# 读取原始数据
df = pd.read_csv("./Data/my_file.csv")

# ========== 任务1：缺失值处理 ==========
df['Rank'] = df['Rank'].fillna(-1).astype('Int64')
df['Peak'] = df['Peak'].fillna(-1)
df['All Time Peak'] = df['All Time Peak'].fillna(-1)
df['Shows'] = df['Shows'].fillna(0).astype('int')
df['Average gross'] = df['Average gross'].fillna('$0')
df['Artist'] = df['Artist'].fillna("Unknown")
df['Tour title'] = df['Tour title'].fillna("Unknown")
df['Year(s)'] = df['Year(s)'].fillna("N/A")

# ========== 任务2：票房金额标准化 ==========
# 清洗 Actual gross 字段
df['gross_notes'] = df.iloc[:, 3].str.extract(r'\[([a-z])\]')  # 提取注释
df['actual_gross_usd'] = (
    df.iloc[:, 3].str.replace(r'\[.*\]', '', regex=True)
                      .str.replace(r'[$,]', '', regex=True)
                      .astype(float)
)

# 清洗 Adjusted gross 字段
df['adjusted_gross_usd'] = (
    df.iloc[:, 4].str.replace(r'\[.*\]', '', regex=True)
                         .str.replace(r'[$,]', '', regex=True)
                         .astype(float)
)

# ========== 任务3：年份拆分与逻辑处理 ==========
df[['start_year', 'end_year']] = df['Year(s)'].str.extract(r'(\d{4})[–-]?(\d{4})?')
df['end_year'] = df['end_year'].fillna(df['start_year'])
df[['start_year', 'end_year']] = df[['start_year', 'end_year']].astype('Int64')

# 逻辑修复：start_year > end_year
swap_mask = df['start_year'] > df['end_year']
df.loc[swap_mask, ['start_year', 'end_year']] = df.loc[swap_mask, ['end_year', 'start_year']].values

# ========== 任务4：巡演信息标准化 ==========
artist_map = {
    'Beyoncé': 'Beyonce',
    'Celine Dion': 'Céline Dion'
}
df['Artist'] = df['Artist'].replace(artist_map)

# 清洗 Tour title，去除特殊符号
df['main_title'] = (
    df['Tour title'].str.replace(r'[†‡\*\[\]a-zA-Z]', '', regex=True)
                    .str.strip()
)

# ========== 任务5：排名字段处理 ==========
# 处理 Peak 字段
df['peak_notes'] = df['Peak'].astype(str).str.extract(r'\[(.*)\]')
df['Peak'] = df['Peak'].astype(str).str.replace(r'\[.*\]', '', regex=True)
df['Peak'] = pd.to_numeric(df['Peak'], errors='coerce').fillna(-1).astype('Int64')

# 处理 All Time Peak 字段（同上）
df['all_time_notes'] = df['All Time Peak'].astype(str).str.extract(r'\[(.*)\]')
df['All Time Peak'] = df['All Time Peak'].astype(str).str.replace(r'\[.*\]', '', regex=True)
df['All Time Peak'] = pd.to_numeric(df['All Time Peak'], errors='coerce').fillna(-1).astype('Int64')

# ========== 任务6：演出场次验证 ==========
# 清洗 Average gross
df['avg_gross_clean'] = (
    df['Average gross'].str.replace(r'\[.*\]', '', regex=True)
                       .str.replace(r'[$,]', '', regex=True)
                       .astype(float)
)

# 计算平均值并比较
df['calculated_avg'] = df['actual_gross_usd'] / df['Shows'].replace(0, np.nan)
df['avg_diff'] = (df['avg_gross_clean'] - df['calculated_avg']).abs()

# 标记异常：超过10%
df['avg_flag'] = (df['avg_diff'] / df['avg_gross_clean']) > 0.10

# ========== 任务7：引用信息处理 ==========
df['ref_notes'] = df['Ref.'].str.extract(r'\[(.*)\]')
df['ref_category'] = df['Ref.'].astype(str).str.extract(r'\[([a-zA-Z])\]')

# ========== 写入清洗后的数据 ==========
df.to_csv('cleaned_music_tours.csv', index=False)
```

## machine learning

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score
from sklearn.model_selection import StratifiedKFold
import joblib

# --- Step 1: 读取数据 ---
train = pd.read_csv('./Data/2/train.csv')
test = pd.read_csv('./Data/2/test.csv')

# 保存测试集 id 以备输出
test_ids = test['id']

# 合并数据用于统一预处理（注意：train有price_range，test没有）
test['price_range'] = np.nan  # 添加空列方便合并
data = pd.concat([train, test], ignore_index=True)

# --- Step 2: 缺失值处理 ---
for col in ['px_height', 'px_width']:
    data[col] = data[col].replace(0, np.nan)  # 将0视为缺失
    data[col].fillna(data[col].mean(), inplace=True)

# 删除 px_height 或 px_width 原始为0（现为nan）后仍然异常的行
data.dropna(subset=['px_height', 'px_width'], inplace=True)

# --- Step 3: 特征工程 ---

# 去除 id
data.drop(columns=['id'], inplace=True)

# 分离出 train/test
train_data = data[~data['price_range'].isna()].copy()
test_data = data[data['price_range'].isna()].copy()

# 提取标签
y = train_data['price_range'].astype(int)
X = train_data.drop(columns=['price_range'])
X_test = test_data.drop(columns=['price_range'])

# 数值特征标准化（包括所有特征）
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_test_scaled = scaler.transform(X_test)

# --- Step 4: 模型训练 + 交叉验证 ---
model = RandomForestClassifier(n_estimators=150, random_state=42)

# 5折交叉验证 F1-score（宏平均）
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
f1_scores = cross_val_score(model, X_scaled, y, cv=cv, scoring='f1_macro')

print(f"5-fold CV Macro F1-score: {f1_scores.mean():.4f}")

# --- Step 5: 模型拟合全训练集 + 保存 ---
model.fit(X_scaled, y)
joblib.dump(model, 'model.pkl')

# --- Step 6: 测试集预测 ---
y_pred = model.predict(X_test_scaled)

# 保存预测结果
submission = pd.DataFrame({
    'id': test_ids,
    'price_range': y_pred.astype(int)
})
submission.to_csv('predictions.csv', index=False)
```

# B卷机器学习

现有一份某地用户的签到信息sign.csv，每条签到信息包含用户ID、经度、纬度、时间戳信息。基于该数据集，使用k-means算法完成用户聚类，数据字段说明如下,输出训练的模型文件，模型文件命名为kmeans_model.pkl；同时提交字符串结果如“k=2,score=0.55”（其中k为聚类簇数，score为轮廓系数）
 
sign.csv数据字段说明
|序号	|字段	|字段描述|
|---|---|---|
|1	|userId	|用户ID|
|2	|latitude	|纬度|
|3	|longitude	|经度|
|4	|timestamp	|时间|


```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import joblib  # 用于保存模型

# 1. 读取签到数据
df = pd.read_csv('./Data/sign.csv')

# 2. 只提取经纬度用于聚类
# 'user_id', , 'timestamp'
X = df[['latitude', 'longitude']].dropna()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. 寻找最佳聚类数 k（一般从2到10尝试）
best_k = 0
best_score = -1
best_model = None

for k in range(2, 11):
    for alg in ['elkan', 'lloyd']:
        # 使用不同的初始化方法和算法
        kmeans = KMeans(n_clusters=k, init='random', algorithm=alg, random_state=42)
        labels = kmeans.fit_predict(X_scaled)
        score = silhouette_score(X_scaled, labels)

        if score > best_score:
            best_k = k
            best_score = score
            best_model = kmeans

# 4. 输出最佳聚类簇数和对应的轮廓系数
print(f'k={best_k},score={best_score:.2f}')

# 5. 保存最佳模型
joblib.dump(best_model, 'kmeans_model.pkl')
```
